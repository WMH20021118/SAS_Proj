```python
orders = pd.read_csv('/data/lab/SAS_Proj/OlistData/olist_orders_dataset.csv')
order_reviews = pd.read_csv('/data/lab/SAS_Proj/OlistData/olist_order_reviews_dataset.csv')
order_items = pd.read_csv('/data/lab/SAS_Proj/OlistData/olist_order_items_dataset.csv')
order_payments = pd.read_csv('/data/lab/SAS_Proj/OlistData/olist_order_payments_dataset.csv')
order_customers = pd.read_csv('/data/lab/SAS_Proj/OlistData/olist_customers_dataset.csv')
products = pd.read_csv('/data/lab/SAS_Proj/OlistData/olist_products_dataset.csv')
sellers = pd.read_csv('/data/lab/SAS_Proj/OlistData/olist_sellers_dataset.csv')
geolocation = pd.read_csv('/data/lab/SAS_Proj/OlistData/olist_geolocation_dataset.csv')
```

```python
# 使用内连接合并所有订单相关数据
orders_complete = pd.merge(orders, order_reviews, on='order_id', how='inner')
orders_complete = pd.merge(orders_complete, order_items, on='order_id', how='inner')
orders_complete = pd.merge(orders_complete, order_payments, on='order_id', how='inner')
orders_complete = pd.merge(orders_complete, order_customers, on='customer_id', how='inner')
orders_complete = pd.merge(orders_complete, products, on='product_id', how='inner')
orders_complete = pd.merge(orders_complete, sellers, on='seller_id', how='inner')
```

```python
orders_complete.to_csv('/data/lab/SAS_Proj/output_file.csv', index=False)

orders_complete
```

```python
import pandas as pd
import matplotlib.pyplot as plt

orders_with_pay = pd.merge(orders, order_payments, on='order_id', how='inner')

# 把order中字符串转化为日期时间数据类型
def transform_datetime(data, column_list):
    for column in column_list:
        data[column] = pd.to_datetime(data[column])

column_list = ['order_purchase_timestamp', 'order_approved_at',
              'order_delivered_carrier_date', 'order_delivered_customer_date', 
              'order_estimated_delivery_date']
transform_datetime(orders_with_pay, column_list)

# 按小时分组并计算每小时的订单总数和支付总额
orders_with_pay['hour'] = orders_with_pay['order_purchase_timestamp'].dt.hour
grouped = orders_with_pay.groupby('hour').agg({
    'order_id': pd.Series.nunique,
    'payment_value': 'sum'
})

# 将支付总额四舍五入到两位小数
grouped['payment_value'] = grouped['payment_value'].round(2)

# 绘图展示每小时的订单总数和支付总额
fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:blue'
ax1.set_xlabel('Hour')
ax1.set_ylabel('Number of Orders', color=color)
ax1.bar(grouped.index, grouped['order_id'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # 实例化一个共享相同x轴的第二个坐标轴
color = 'tab:red'
ax2.set_ylabel('Total Payment Value', color=color)
ax2.plot(grouped.index, grouped['payment_value'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # 调整布局以防止重叠
plt.title('Number of Orders and Total Payment Value per Hour')
plt.xticks(grouped.index)
plt.show()
```
